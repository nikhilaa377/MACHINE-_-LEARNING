- `since our Training data Actual Points and Prediction Points are very close to Each other we are getting Best Training Accuracy and Best Training Loss `

- `What if we have Lot of Difference Between Actual Point and Prediction Point `

- `we Need to take Help of Gradient Descent Optimizer -> which is used to reduce the Loss and Increase Accuracy `

- `Gradient Descent works with Cost Function | Step Size | New Slope and New Intercept `

- `Lets Assume some data which is having lot of difference between acutal point and prediction point -> How Gradient Descent works:`

```python
def gra(x1,y1):
    # slope and intercept
    m = c = 0
    n = len(x1)
    lr = 0.01
    plt.scatter(x1,y1,color='red',marker='+',linewidth=5)
    # Model
    for i in range(100):
        y_predictions = m * x1 + c
        cost_fun = (1/n) * sum([k**2 for k in (y1 - y_predictions)])
        acc = r2_score(y1,y_predictions)
        plt.plot(x1,y_predictions,color='red',linewidth=1)
        #print('m = ',m,'c = ',c , 'itera = ',i , 'cost fun = ',cost_fun , 'acc=',acc)
        print('cost fun = ',cost_fun , 'acc=',acc)
        # loss function / cost function
        # formula MSE = 1/n * (y - mx+c)**2
        # derivative m = -(2/n)*sum(x*(y-y_predicted))
        # derivative c = -(2/n)*sum(y-y_predicted)
        md = -(2/n)*sum(x1*(y1-y_predictions))
        cd = -(2/n)*sum(y1-y_predictions)

        # update step_size = md * lr , cd * lr
        # update slop  and c
        m = m - md * lr
        c = c - cd * lr
```

```python
inde = np.array([10,20,40,30,60,50]) # X_train
dep = np.array([4,14,32,21,55,44]) # y_train
import warnings
warnings.filterwarnings('ignore')

gra(inde,dep)
```

- `This Technique will be used when we have lot of difference between actual point and predicton point | if there is no difference no need to implement Gradient Descent.`
